```python
x = range(1,11,1)
cxMarks= [8,9,6,3,5,7,4,1,2,6]
cyMarks= [5,8,9,6,3,2,4,8,8,9]


# write a code to generate below graph
plt.plot(x,cxMarks, color='r', ls=':', marker='o')
plt.plot(x,cyMarks, color='b', ls='--',  marker='p')
plt.xlabel("Roll-Number")
plt.ylabel("CPI")
plt.title("Line_Graph", fontsize=20)
plt.annotate(xy=[8,1],text="Lowest CPI :(")
# plt.annotate(xy=[2,9],text="Highest CPI")
plt.legend(["cx", "cy"], loc=3)      #loc --> Quadrant;
plt.show()
```

### 01) WAP to demonstrate the use of Pie chart.

```python
x=[5000,3000,3000,2000]
y=["Anime","Movie","Web_Series","TV_Show"]
c=["r","y","g","b"]
ex=[0.5, 0.0, 0.0, 0.0]
plt.pie(x,labels=y,colors=c, radius=1.5, shadow=True, explode=ex, autopct="%0.2f%%", labeldistance=1.2, textprops={"fontsize":20})
plt.legend(y,title ="Reference", fontsize=15, bbox_to_anchor =(1.27, 0.8, 0.8, 1))
# plt.title('expenses', color='b', )
plt.show()
```

### 03) WAP to demonstrate the use of Bar chart.

```python
x=["PYTHON","C","C++","JAVA"]
y=[85,70,60,90]
c=["r","g","b","m"]
plt.bar(x,y,color=c,align='edge', edgecolor='c', alpha=0.8)
plt.xlabel("Languages")
plt.ylabel("Rating")
plt.title("Popularity Of Programming language")
plt.show()
```

### 01) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i])
```

```python
x=["PYTHON","C","C++","JAVA"]
y=[85,70,60,90]
c=["r","g","b","m"]
plt.bar(x,y,color=c,align='center', edgecolor='c', alpha=0.8)
plt.xlabel("Languages")
plt.ylabel("Rating")
plt.title("Popularity Of Programming language")
addlabels(x,y)
plt.show()
```

### 02) WAP create a Scatter Plot with several colors in Matplotlib?

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
c=["r","g","b","m", "c"]
size=[100,200,300,500,400]
plt.scatter(x,y,color=c, marker="*", s=size)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter_Plot", fontsize=20)
# plt.show()
plt.savefig("myChart.jpg")
```
